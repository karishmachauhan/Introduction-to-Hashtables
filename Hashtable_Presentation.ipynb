```python
# Importing libraries

from lolviz import *
import numpy as np
import pandas as pd
```

```python
# Creating random numpy array

# Size of sample size

n = 5000

# Random generating integers from 0 to 10000 of size 5000

A = list(np.random.randint(low = 0, high = 10000, size = n)) 

# Picking random element from the list

search_element = A[-10]

print(search_element)
```

```python
# Linear Search Function

# lsearch is the name of the function and A is the list and x is the target search element

def lsearch(A, search_element):

  # Looping through list to find the element x

  for row in A:

    # Condition if any element in A 

    if row == search_element:

      # The function returns True that means, search element is present in the list_A

      return True 


  return False
```

```python
# Let's Create a hash function 

def hash(x):

    return x % 1000
```

```python
# Make sure each bucket is a separate list

buckets = [[] for i in range(1000)] 
```

```python
# Filling up buckets

for a in A:

    index = hash(a)

    values = buckets[index]

    if (values == None):

      values = []
      
    values.append(a)

    buckets[index] = values
```

```python
# Hash search Function which returns true if element is present

def hsearch(hashtable, search_element):

    index = hash(search_element)

    values = hashtable[index]

    for value in values:

      if (value == search_element):

        return True


    return False
```

```python
# Time taken by linear search to search an element

# Using inbuilt notebook functionality to time the computation

%time lsearch(A, search_element)
```

```python
# Time taken by Hash search to search an element

%time hsearch(buckets, search_element)
```

###**We traded off computation complexity for time**

*   Linear Search : 
        • Lesser Complexity
        • More Time
     
*   Hashtable Search :
        • More Complexity
        • Less Time

```python
# Let's create a new hash function for strings

def hash_str(o):

    h = 0

    for i in range(len(o)):

        h = h*31 + ord(o[i])
  

    return h
```

```python
# Creat the look up function for linear search

def llookup(A, x):

    for k, v in A:

        if k == x:

            return v

            
    return None
```

```python
# Creat the look up function for hashtable search

def hlookup(A, x):

  index = hash_str(x) % 7

  for k, v in hbuckets[index]:

    if k == x:

      return v

      
  return None
```

```python
# Building hash tables using list of sets (Key, Value) Pairs

# Creating empty buckets for hash table

hbuckets = [[] for i in range(0, 20)]

for k,v in companies_priority:

    index = hash_str(k) % 20

    hbuckets[index].append((k, v))
```
